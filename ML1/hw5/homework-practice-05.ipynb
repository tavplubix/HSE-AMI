```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

```python
# Пример:
from sklearn.linear_model import LinearRegression
X, y = datasets[2]
lr  = LinearRegression().fit(X, y)
plot_surface(lr, X, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
def task1(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    dtc = DecisionTreeClassifier(random_state=42)
    dtc.fit(X_train, y_train)
    pred = dtc.predict(X_test)
    print('accuracy:', accuracy_score(y_test, pred))
    plt.figure(figsize=(15, 4))
    plt.subplot(1, 3, 1)
    plt.title('train')
    plot_surface(dtc, X_train, y_train)
    plt.subplot(1, 3, 2)
    plt.title('test')
    plot_surface(dtc, X_test, y_test)
    plt.subplot(1, 3, 3)
    plt.title('all')
    plot_surface(dtc, X, y)
    plt.show()
```

В ней признаки объекта записаны в первых пяти столбцах, а в последнем записана целевая переменная (класс: 0 или 1). Постройте на одном изображении пять кривых "порог — значение критерия Джини" для всех пяти признаков. Отдельно визуализируйте scatter-графики "значение признака — класс" для всех пяти признаков.

```python
def find_best_split(feature_vector, target_vector):
    _tmp = np.array([feature_vector, target_vector])
    _tmp = _tmp[:, np.argsort(_tmp[0])]
    lt = _tmp[0][:-1]
    rt = _tmp[0][1:]
    t = (lt + rt) * 0.5

    p1l = np.cumsum(_tmp[1])[:-1]
    p1r = (_tmp[1] == 1).sum() - p1l
    Rl = np.array(range(1, _tmp.shape[1]))
    p0l = Rl - p1l
    p0r = (_tmp[1] == 0).sum() - p0l

    Hl = 1 - (p0l / Rl) ** 2 - (p1l / Rl) ** 2
    Rr = ((_tmp.shape[1]) - np.array(range(1, _tmp.shape[1])))
    Hr = 1 - (p0r / Rr) ** 2 - (p1r / Rr) ** 2

    Q = - (Rl * Hl + Rr * Hr) / _tmp.shape[1]
    tr, idx = np.unique(t, return_index=True)
    ginis = Q[idx]
    best_idx = np.argmax(ginis)

    return tr, ginis, tr[best_idx], ginis[best_idx]
```

```python
import pandas as pd
X = pd.read_csv('students.csv')
fig, ax = plt.subplots()
plt.title('Зависимость значения критерия Джини от порога')
plt.xlabel('порог')
plt.ylabel('значение критерия Джини')
for f in ('STG', 'SCG', 'STR', 'LPR', 'PEG'):
    # тот, кто добавил пробел в ' UNS', бeдет гореть в аду, потому что ВНЕЗАПНО ' UNS' != 'UNS'
    tr, g, _, _ = find_best_split(X[f], X[' UNS'])
    ax.plot(tr, g, label=f)
ax.legend(loc='upper right', shadow=True, fontsize='x-large')
```
